```python
import numpy as np

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

AND  = np.array([0,0,0,1])
OR   = np.array([0,1,1,1])
NAND = np.array([1,1,1,0])
NOR  = np.array([1,0,0,0])

class Perceptron:

    def __init__(self, lr=0.1, epochs=10):
        self.lr = lr
        self.epochs = epochs

    def activation(self, z):
        return 1 if z >= 0 else 0

    def fit(self, X, y):

        n_features = X.shape[1]

        
        self.w = np.zeros(n_features)
        self.b = 0

        
        for _ in range(self.epochs):

            for xi, target in zip(X,y):

                z = np.dot(xi,self.w) + self.b
                y_pred = self.activation(z)

                error = target - y_pred

            
                self.w += self.lr * error * xi
                self.b += self.lr * error

    def predict(self,X):

        outputs = []
        for xi in X:
            z = np.dot(xi,self.w) + self.b
            outputs.append(self.activation(z))

        return np.array(outputs)



def train_gate(name, y):

    model = Perceptron(lr=0.1, epochs=10)
    model.fit(X,y)

    predictions = model.predict(X)

    print("\n======================")
    print(name,"Gate")
    print("======================")

    print("Weights:",model.w)
    print("Bias:",model.b)

    print("\nPredictions")

    for i in range(len(X)):
        print(X[i],"->",predictions[i])



train_gate("AND",AND)
train_gate("OR",OR)
train_gate("NAND",NAND)
train_gate("NOR",NOR)
```
